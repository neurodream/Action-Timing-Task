```python
from BFTUSk.analysis import *
from BFTUSk.sim_task import *

dfs_random = []
dfs_early = []
dfs_middle = []
dfs_late = []

for i in tqdm(range(10)):
    dfs_random.append(run_sim(exploit_kill=False))
    dfs_early.append(run_sim(strategy="early", exploit_kill=False))
    dfs_middle.append(run_sim(strategy="middle", exploit_kill=False))
    dfs_late.append(run_sim(strategy="late", exploit_kill=False))

print(dfs_random[0].columns.tolist())
```

```python
import matplotlib.patches as mpatches

rand = mpatches.Patch(color='grey', label='random')
early = mpatches.Patch(color='green', label='early')
mid = mpatches.Patch(color='orange', label='middle')
late = mpatches.Patch(color='darkred', label='late')

f, axes = plt.subplots(2,1, sharex=True)
for df in dfs_random:
    axes[0].plot(sorted(df["rt_frames"]), color="grey")
    axes[1].plot(sorted(df["score"]), color="grey")
for df in dfs_early:
    axes[0].plot(sorted(df["rt_frames"]), color="green")
    axes[1].plot(sorted(df["score"]), color="green")
for df in dfs_middle:
    axes[0].plot(sorted(df["rt_frames"]), color="orange")
    axes[1].plot(sorted(df["score"]), color="orange")
for df in dfs_late:
    axes[0].plot(sorted(df["rt_frames"]), color="darkred")
    axes[1].plot(sorted(df["score"]), color="darkred")
axes[0].set_ylabel("reaction time (frames)\n(sorted)")
axes[1].set_ylabel("score")
axes[0].set_title("independent var")
axes[1].set_title("dependent var")
plt.xlabel("trial N")
axes[0].legend(handles=[rand, early, mid, late], title="response strategy")
plt.tight_layout()
plt.show()
```
