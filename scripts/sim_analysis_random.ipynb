```python
from BFTUSk.analysis import *
from BFTUSk.sim_task import *

dfs = []

for i in range(10):
    dfs.append(run_sim())

print(dfs[0].columns.tolist())
```

```python
plt.figure()
for df in dfs:
    plt.plot(sorted(df["rt_frames"]))
plt.show()
```

```python
plt.figure()
for df in dfs:
    cum_reward = np.cumsum([val*won for val, won in zip(df["reward_size"], df["reward"])])
    plt.plot(cum_reward)
plt.show()
```
